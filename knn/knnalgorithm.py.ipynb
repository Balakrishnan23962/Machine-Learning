```python
from sklearn.datasets import make_blobs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
inputs,targets =make_blobs(n_samples=1000,random_state=365,centers=[(-3,3),(0,0),(3,4)])
data=pd.DataFrame(data=inputs,columns=["Feature1","feature2"])
data["target"]=targets
knn=sn.color_palette(["red","blue","black"])
x_t,x_tt,y_t,y_tt=train_test_split(inputs,targets,test_size=0.2,random_state=365,stratify = targets)
clf=KNeighborsClassifier(n_neighbors=3)
clf.fit(x_t,y_t)
f1=0.4
f2=2.4
clf.predict([[f1,f2]])
neighbors=clf.kneighbors([[f1,f2]])
sn.set()
plt.figure(figsize=(10,9))
sn.scatterplot(x=x_t[:,0],y=x_t[:,1],hue=y_t,palette=knn,markers=[',','^','p'],style=y_t,s=100,legend=True)
sn.scatterplot(x=[f1],y=[f2],style=[f2],markers=["o"],s=100,legend=False)
px=[]
py=[]
for i in neighbors[1]:
    px.append(x_t[i,0])
    py.append(x_t[i,1])
plt.scatter(px,py,s=200,edgecolors="y")
```
